# 5. Compute Derived Variables

```python
# Calculate waist-to-hip ratio (WHR)
df_cluster["WHR_i0"] = df_cluster["waist_circumference_i0"] / df_cluster["hip_circumference_i0"]
df_cluster[["participant_ID","waist_circumference_i0", "hip_circumference_i0", "WHR_i0"]].head()
```

```python
# Average blood pressure across two measurements (ignoring NaN)
df_cluster.loc[:, "DBP_i0"] = df_cluster[["DBP_i0_a0", "DBP_i0_a1"]].mean(axis=1, skipna=True)
df_cluster.loc[:, "SBP_i0"] = df_cluster[["SBP_i0_a0", "SBP_i0_a1"]].mean(axis=1, skipna=True)
df_cluster.head()
```

```python
# Encode current smoking as binary (1 = current smoker, 0 = otherwise)
df_cluster["current_smoking_i0"] = (df_cluster["smoking_i0_label"] == "Current").astype(int)
df_cluster["current_smoking_i0"].value_counts()
```
